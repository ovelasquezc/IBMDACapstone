# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
QUERY = """
SELECT ConvertedComp,Age,WorkWeekHrs,CodeRevHrs,MainBranch
FROM master
"""
df = pd.read_sql_query(QUERY,conn)

import matplotlib.pyplot as plt
ax = df['ConvertedComp'].plot(kind='hist', density=True, alpha=0.6, bins=30, label='Histogram')
df['ConvertedComp'].plot(kind='kde', ax=ax, label='Density')

ax.set_xlabel('ConvertedComp')
ax.set_ylabel('Density')
ax.set_title('Distribution Curve of ConvertedComp')
plt.legend()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
df.boxplot(column='Age')

# Adding labels and title for better understanding
#plt.xlabel('Age')
#plt.ylabel('Values')
plt.title('Box Plot of Age')

# Display the plot
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
# Creating the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['WorkWeekHrs'], alpha=0.6)

# Adding labels and title for better understanding
plt.xlabel('Age')
plt.ylabel('Hours')
plt.title('Age vs WorkWeek Hours')

# Display the plot
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
plt.figure(figsize=(10, 6))
bubble_size = df['Age']
plt.scatter(df['WorkWeekHrs'], df['CodeRevHrs'], s=bubble_size, alpha=0.5, c=bubble_size, cmap='viridis')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.title('Bubble Plot of WorkWeekHrs vs CodeRevHrs')
plt.colorbar(label='Age')

# Display the plot
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
QUERY = """
SELECT DatabaseDesireNextYEar as Database
FROM DatabaseDesireNextYEar WHERE 
DatabaseDesireNextYear in
(SELECT DatabaseDesireNextYear FROM
(SELECT DatabaseDesireNextYear,COUNT(*) as Respondents
FROM DatabaseDesireNextYEar
GROUP BY DatabaseDesireNextYear
ORDER BY Respondents DESC
LIMIT 5))
"""

dfb = pd.read_sql_query(QUERY,conn)

# Count the occurrences of each category
category_counts = dfb['Database'].value_counts()

# Creating the pie chart
plt.figure(figsize=(8, 8))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%')

# Adding a title
plt.title('Distribution of Categories')

# Display the plot
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
filtered_df = df[(df['Age'] >= 30) & (df['Age'] <= 35)]

overall_median_workweekhrs = filtered_df['WorkWeekHrs'].median()
overall_median_coderevhrs = filtered_df['CodeRevHrs'].median()

plt.figure(figsize=(10, 6))

plt.bar(['Age Group'], [overall_median_workweekhrs], color='blue', label='WorkWeekHrs')
plt.bar(['Age Group'], [overall_median_coderevhrs], color='green', bottom=overall_median_workweekhrs, label='CodeRevHrs')

plt.text('Age Group', overall_median_workweekhrs / 2, f'{overall_median_workweekhrs:.1f}', ha='center', va='center', color='white', fontsize=10)
plt.text('Age Group', overall_median_workweekhrs + overall_median_coderevhrs / 2, f'{overall_median_coderevhrs:.1f}', ha='center', va='center', color='white', fontsize=10)

#plt.xlabel('Age Group')
plt.ylabel('Median Hours')
plt.title('Overall Median WorkWeekHrs and CodeRevHrs for Ages 30-35')
plt.legend()

# Display the plot
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

# Filter the DataFrame for Age between 45 and 60
filtered_df = df[(df['Age'] >= 45) & (df['Age'] <= 60)]

# Group by Age and calculate median ConvertedComp
grouped_df = filtered_df.groupby('Age').median()

# Plot the line chart
plt.figure(figsize=(10, 6))
plt.plot(grouped_df.index, grouped_df['ConvertedComp'], marker='o', color='b', linestyle='-')

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.title('Median ConvertedComp for Ages 45-60')
plt.grid(True)

# Display the plot
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
branch_counts = df['MainBranch'].value_counts()

plt.figure(figsize=(10, 6))
branch_counts.plot(kind='barh', color='green')
plt.xlabel('Count')
plt.ylabel('Main Branch')
plt.title('Count of Main Branch Categories')
plt.grid(axis='x')  # Add grid lines on the x-axis for better readability
plt.show()
```
